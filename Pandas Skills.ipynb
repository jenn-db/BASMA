# Group By and Aggregate Functions

```python
# getting the results by day
group_by_date = df2.groupby('Day')
group_by_date.head()
```

```python
# average results by day
average_results = group_by_date.mean()
average_results.head()
```

```python
# counting the rows that were aggregated 
counting_rows = df2.groupby('Day').count()
counting_rows.head()
```

```python
# highest results per day and platform 
highest_results = df2.groupby(['Day', 'Platform']).max()
highest_results.head()

# if we don't add the platform, it will add the platform based on alphabetical hierarchy
# we can use the same thing for minimum results - .min - 
```

```python
# total values per day
total_values = df2.groupby('Day').sum()
total_values.head()
```

```python
# calculating the max and total amount spent, and Unique Purchases per day and platform
agg_f = df2.groupby(['Day','Platform']).agg({'Amount Spent (USD)': ['max', 'sum'], 'Unique Purchases': ['max', 'sum']})
agg_f.head()
```

```python
# we can also use describe for high level overview of the aggregations
general_ov = df2.groupby('Day').describe()
general_ov.head()
```
